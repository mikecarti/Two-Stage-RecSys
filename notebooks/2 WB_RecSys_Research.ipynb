# Отбор Кандидатов

## Коллаборативная фильтрация (CF):
(memory-based, model-based)
### Плюсы:
Не требует никакой информации о пользователях или предметах, кроме взаимодействия пользователя с предметом.
Может улавливать сложные закономерности в поведении пользователей.
Этот подход прост и быстр для получения результатов
на основе небольшого числа наблюдений.
 Не требует учета пользовательских оценок, которые
которые требуют коллаборативные и основанные на контенте методы.
### Минусы:
Проблема "холодного старта": невозможно рекомендовать товары новым пользователям или товары без каких-либо взаимодействий.
Неразборчивость: Может страдать от нехватки данных, если пользователи взаимодействуют только с несколькими предметами.
Сбор всей информации о пользователе нецелесообразен для
эффективной демографической фильтрации, учитывая вопросы безопасности
и проблемы конфиденциальности.



## Матричная факторизация (SVD):

### Плюсы:
Справляется с проблемой разреженности лучше, чем традиционная CF, за счет уменьшения размерности матрицы взаимодействия пользователя и предмета.
Улавливает латентные факторы, лежащие в основе предпочтений пользователей и характеристик предметов.
### Минусы:
По-прежнему страдает от проблемы "холодного старта".
Может потребоваться настройка гиперпараметров, таких как количество латентных факторов.


## Фильтрация на основе содержимого (Content Based):

### Плюсы:
Не полагается на взаимодействие пользователя с элементами и в некоторой степени решает проблему "холодного старта".
Использует характеристики элементов для составления рекомендаций, что может быть полезно при наличии информации об элементах.

Методы, основанные на содержании, должны только тщательно изучать предметы и профиль пользователя для получения рекомендаций. Метод, основанный на содержании, может информировать пользователя о характеристиках, на которых основывается рекомендация. Новые товары могут быть рекомендованы до того, как их оценит большое количество пользователей.
### Минусы:
Ограничен рекомендациями товаров, похожих на те, с которыми пользователь уже взаимодействовал, что потенциально может привести к отсутствию разнообразия в рекомендациях.
Требуется извлечение и разработка характеристик, что может быть затруднительно при отсутствии соответствующих метаданных об объекте.

Если содержание не содержит достаточного количества
информации для точной дифференциации предметов, то
полученные рекомендации не будут точными.
Система не предоставляет надежных рекомендаций, так как
нет достаточного количества информации для построения
хорошего профиля пользователя.
Еще одним недостатком является проблема "случайности", когда
система будет рекомендовать или предлагать те предметы, которые
имеют высокую популярность, но низкий serendipity



## Гибридные рекомендательные системы (CF + Content-Based):

### Плюсы:
Объединяет сильные стороны нескольких рекомендательных подходов для обеспечения более точных и разнообразных рекомендаций.
Может смягчить недостатки отдельных методов.
Проблема холодного старта из-за коротких профилей пользователей и
Наличие редких оценок пользователя может быть эффективно решено
эффективно с помощью гибридной рекомендательной системы
### Минусы:
Повышенная сложность реализации и обслуживания.
Требуется тщательная интеграция различных компонентов.
Разработка гибридной рекомендательной системы является сложной задачей
Разработка всех возможных комбинаций гибридной рекомендательной системы также не представляется возможной.



## Рекомендательные системы с учетом времени (Time dependent Recommender Systems):

### Плюсы:
Учитывают временную информацию при составлении рекомендаций, что может быть полезно для учета изменений в предпочтениях пользователей с течением времени.
Может учитывать сезонность и тенденции в поведении пользователей.
Временная погрешность: интересы целой группы пользователей могут меняться с течением времени. Например, если вы рекомендуете музыку и появляется новый музыкальный стиль (скажем, trap music), ваша модель должна уметь его учитывать.
Смещение пользовательской предвзятости: пользователь может изменить свои привычки оценивания с течением времени
Предвзятое отношение к предмету: Популярность товара меняется со временем
Смещение предпочтений пользователя: вкусы пользователя меняются

### Минусы:
Сложность моделирования временной динамики и учета времени в алгоритмах рекомендаций.
Требуется достаточное количество исторических данных для выявления значимых временных закономерностей.

#### Simple Association
$$
s_{AR}(a,b) = \sum_{s \in S} \#(a,s) * \#(b,s)
$$
the score is the number of co-occurrences (взаимо-вхождений) of the two items in a sequence. $\#(a,s) - $ кол-во раз, которое товар $a$ появляется в последовательности $s$.

#### Markov Chains
$$
S_{MC}(a,b) = \sum_{s \in S} \sum_{i=1}^{|s|-1} I(a=s_i) * I(b=s_{i+1} )
$$
the score is the number of co-occurrences of the two **consecutive** items in a sequence.

## Методы Соседства (I-KNN, S-KNN)

### Pros of Neighborhood Methods:

Они используют сходство между пользователями или предметами, основанное на их взаимодействии, что позволяет выявить локальные закономерности в данных.
Они относительно просты в реализации и понимании по сравнению с более сложными моделями.
Они могут давать неожиданные рекомендации, выявляя пользователей или предметы с похожим поведением. I-KNN также работает с последовательностями как и рек. системы с учетом времени.
### Cons of Neighborhood Methods:

Они подвержены проблеме разреженности, особенно в наборах данных с небольшим количеством взаимодействий с пользователем или товаром.
Они могут плохо масштабироваться на большие массивы данных из-за требований к вычислительной мощности и памяти.
Они могут страдать от проблемы "холодного старта", особенно для новых пользователей или предметов без достаточной истории взаимодействий.

## Рекомендательные системы на основе глубинного обучения:

### Плюсы:
Способны автоматически изучать сложные закономерности на основе данных, что потенциально может привести к более точным рекомендациям.
Могут работать с большими массивами данных и включать в себя различные типы данных (например, поведение пользователя, характеристики товара).
### Минусы:
Требует больших объемов данных для обучения.
При отсутствии надлежащей регуляризации может страдать от чрезмерной подгонки.



### Neural Collaborative Filtering
### Плюсы:
- SOTA, использует Матричную Факторизацию и Глубинное Обучение, чтобы улучшить результаты классической матричной факторизации.


### Минусы:
- был назван SOTA по-читерски: Стивен Ренделл разоблачил авторов. Сравнивали это с матричным разложением, которое не сошлось до конца, если учить нормально скалярное проиведение, оно побивает NCF.
- Не сходится к нулевой ошибке для идеально подобранных данных (нейронная сеть не может аппроксимировать скалярное произведение)


### BERT4Rec

- Bidirectional self-attention to model user behavior sequences. ("unidirectional models
restrict the power of hidden representation for items in the historical sequences"). То есть модель не предполагает строгой последовательности покупок.

# Ранжирование

## RankNet
Плюсы:

RankNet - это нейросетевой подход к обучению ранжированию. Он позволяет гибко моделировать сложные взаимосвязи между признаками и релевантностью. Он может эффективно обрабатывать большие пространства признаков.

Минусы:

 При отсутствии надлежащей регуляризации может страдать от чрезмерной подгонки. Обучение может быть вычислительно дорогим, особенно для больших наборов данных.


- Предполагает, что все оценки релевантности известны

## LambdaRank

Плюсы:LambdaRank оптимизирует непосредственно для метрики ранжирования (например, NDCG или MAP), что может привести к лучшей производительности с точки зрения метрики оценки.Он учитывает тот факт, что не все парные различия одинаково важны.

Минусы: Требуются парные обучающие данные, получение которых может быть дорогостоящим. Процесс обучения может быть более медленным по сравнению с другими методами.

## LambdaMART

Плюсы:

LambdaMART расширяет LambdaRank за счет использования методов градиентного усиления, в частности MART (Multiple Additive Regression Trees).Это позволяет создавать более гибкие и мощные модели.


Минусы:

Как и LambdaRank, LambdaMART требует парных обучающих данных и может иметь более медленное время обучения по сравнению с более простыми методами.

# Soft Rank
Плюсы:

Soft Rank - это дифференцируемая аппроксимация ранговых функций потерь, что делает ее подходящей для сквозного обучения нейронных сетей.Он обеспечивает плавный ландшафт оптимизации.

Минусы:

Soft Rank не может напрямую оптимизировать саму метрику ранжирования, что может привести к неоптимальной производительности по сравнению с такими методами, как LambdaRank или LambdaMART.

## LambdaLoss
Все остальные списочные методы (RankNet, LambdaRank, SoftRank, ListNet, ...) являются специальными конфигурациями этой общей схемы.

Действительно, их потери получаются при точном выборе вероятности p(y | s, π) и распределения ранжированного списка p(π | s).

Этот фреймворк позволяет нам определять функции потерь, непосредственно связанные с метриками ранжирования, которые мы хотим оптимизировать. Это позволяет значительно улучшить современное состояние задач Learning to Rank.

Плюсы: LambdaLoss - это фреймворк, который обобщает LambdaRank и позволяет включать различные функции потерь, помимо парных различий, обеспечивая большую гибкость в моделировании.


Минусы: Как и другие методы, основанные на LambdaRank, LambdaLoss требует парных обучающих данных и может иметь более медленное время обучения. Также может потребоваться тщательная настройка гиперпараметров.

## ListNet
ListNet:
Плюсы:

ListNet - это основанный на нейронных сетях метод обучения ранжированию, который напрямую оптимизирует функцию потерь по списку, такую как потери от перестановки. Он обеспечивает принципиальный подход к оптимизации ранжирования непосредственно на основе всего списка документов, а не парных предпочтений.

Плюсы: Он может эффективно работать со списками переменной длины.

Минусы: Обучение ListNet может быть вычислительно дорогим, особенно для больших наборов данных, из-за использования нейронных сетей. Для достижения оптимальной производительности может потребоваться тщательная настройка гиперпараметров.

Дополнительное замечание: было показано, что ListNet демонстрирует конкурентоспособную производительность по сравнению с другими современными методами обучения ранжированию, особенно в сценариях, когда для обучения доступен весь список документов.

# Итоги

Проанализировав набор моделей подбора кандидатов, я решил остановится на BERT4Rec.

Этот метод эффективно вбирает в себя исторические данные, с помощью bidirectional self-attention. Мне кажется, что данных достаточно много, а значит модель, которая базируется на BERT'e будет работать неплохо. Так как у нас есть временные метки, надо этим пользоваться. Остальные модели либо не учитывают временную составляющую, либо просто выглядят не так серьезно и оснавательно. Например Time-Dependent Reccomender Systems не хватает latent обучаемых параметров, а также такие системы, в отличии от BERT4Rec ожидают строгой последовательности, когда BERT4Rec позволяет последовательностям идти в нестрогом порядке.

Относительно ранжирующих методов, я решил попробовать несколько методов, потому что на данный момент тяжело оценить какой будет иметь лучшие показатели.
